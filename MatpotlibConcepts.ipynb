```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x=(1,2,3)
```

```python
y=(1,4,10)
```

```python
x
```

```python
y
```

```python
#There are 2 ways of creating plots. One is functional method and two is OOM
#Below is functional method
plt.plot(x,y)
```

```python
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.plot(x,y)
plt.show()#same as printing plt
```

```python
#Creating multi-plots on the same canvas
plt.subplot(1,2,1)#args here are no. of rows, no. of columns, plot number
#now plot something here
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented Method - create figure objects and call methods on it
fig=plt.figure()#This has created a blank figure object. Let us call methods on it
#add_axes is a method on figure object and we pass a list of args to it (left, bottom, width, height)
#Note that, left, bottom, width and height values should always be bw 0 and 1
ax = fig.add_axes([0.1,0.1,0.8,0.7])
ax.plot(x,y)
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Plot using OOM')
```

```python
#The above OOM is very similar to what we did with the plt.plot earlier. It is just that we have more control here
#over what we are plotting using the figure object
```

```python
#Now let us see how to put multiple plots on same canvas using OOM
fig = plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
#Below we are creating a plot within another plot, by specifying values of axes2 as 20% left, 50% bottom, 50% width and
#30% height of that of axes1
```

```python
#Let us move axes2 to bottom corner of axes1
fig = plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.5,0.15,0.4,0.3])
axes1.plot(x,y)
axes2.plot(x,y)
```

```python
fig,axes=plt.subplots()#Here we are fetching both fig and axes objects in one shot
#plt.subplots is essentially same as getting fig object first and then getting axes object (by the call fig.add_axes)
#in one step 
axes.plot(x,y)
```
